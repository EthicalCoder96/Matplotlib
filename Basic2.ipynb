## **Chapter-1**
#### Fundamentals of matplotlib

```python
# importing libraries
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('classic')
# plt.style.use('default')
```

#### **Important Note**
##### ***When to use 2-D graph ?***
##### ***1] when data is bivariate***
##### ***2] first column is numerical and second column is categorical***
##### ***3] used in time series data***

```python
# Plotting simple graph
price = [48000,51000,49000,50000,48000]
years = [2019,2020,2021,2022,2023]
plt.plot(years,price)
plt.grid()
plt.show()
```

```python
# Creating using pandas + matplotlib
batsman = pd.read_csv('sharma-kohli.csv')
print(batsman)
```

```python
# Plotting Virat Kohli's graph
plt.plot(batsman['index'],batsman['V Kohli'],lw=2,marker='o')
plt.title("Virat Kohli's Runs")
plt.xlabel('Season')
plt.ylabel('runs')
plt.grid()
plt.show()
```

```python
# Plotting sharma and kohli's record simultaneously
plt.plot(batsman['index'],batsman['V Kohli'],marker='o',color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],marker='o',color='blue')
plt.grid(ls='-')
plt.show()
```

```python
# Plotting sharma and kohli's record legend
plt.plot(batsman['index'],batsman['V Kohli'],marker='o',color='red',label='RCB')
plt.plot(batsman['index'],batsman['RG Sharma'],marker='o',color='blue',label='MI')
plt.grid(ls='-')
plt.legend(loc="best")
# upper right
# upper left ....etc .
plt.show()
```

```python
# Limiting axis
price = [48000,51000,49000,50000,48000,4500000]
years = [2019,2020,2021,2022,2023,2024]
plt.plot(years,price,marker='o')
plt.grid(ls='-',color='k')
# To remove this outlier we use lim function
plt.ylim(0,100000)
# Similar for x axis
# do this when you have outlier
plt.show()
```

## **Chapter-2**
#### Fundamentals of scatter plot

#### **Important Note**
##### ***When to use scatter plot ?***
##### ***1] when data is bivariate***
##### ***2] first column is numerical and second column is numerical***
##### ***3] used to find corelation between two points***

```python
x = np.linspace(-10,10,50) # 50 values from -10 to 10
y = 10*x + 3 + np.random.randint(0,300,50) # 50 values from 0 to 300 by adding 10*x + 3
# print(x)
# print(y)
```

```python
plt.scatter(x,y)
plt.grid()
plt.show()
```

```python
# Creating using pandas and matplotlib
batterDataset = pd.read_csv('batter.csv')
print(batterDataset)
```

```python
# Selecting first 50 batsman
first50 = batterDataset.head(50)
print(first50)
```

```python
# Plot scatter graph
plt.scatter(first50['avg'],first50['strike_rate'],s=(first50['runs']/50))
plt.grid()
plt.title('Average vs Strike Rate')
plt.xlabel('Average')
plt.ylabel('Strike Rate')
plt.show()
```

#### ***scatter vs plot***
##### **plot is faster while scatter is slower**
