## Color and Line reference
##### While plotting you have to mention type of marker , line , color
##### there are some references for colour and line
#### *For line*
###### 1] - for solid line
###### 2] -- for dashed line
###### 3] -. for dashed dotted line
###### 4] : for dotted line
#### *For colour*
###### 1] r for red
###### 2] b for blue
###### 3] g for green 
###### 4] y for yellow
###### 5] c for cyan
###### 6] m for magenta
###### 7] k for black
###### 8] w for white

```python
plt.plot(xPoints,"o-.r")
plt.show()
```

#### **Marker Size**
###### just put ms = value in .plot()

```python
plt.plot(xPoints,"o:",ms=10)
plt.show()
```

#### **Marker Edge Colour(mec)**
###### changing marker edge colour 
###### use mec in .plot()
###### edge colours of circle become black

```python
plt.plot(xPoints,"o-.",ms=20,mec='k')
plt.show()
```

#### **MARKER FACE COLOUR(mfc)**
###### use mfc in .plot()

```python
plt.plot(xPoints,"*-.",ms=20,mec='r',mfc='y')
plt.show()
```

##### **Changing colours and sizes**

```python
import matplotlib.pyplot as plt
import numpy as np 

X7 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
Y7 = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7])
colors = np.array([0,10,20,30,40,45,50,55,60,70,80,90,100])
sizes = np.array([20,50,100,200,500,1000,60,60,90,10,300,600,800])
plt.scatter(X7,Y7,cmap='viridis',c = colors,s=sizes)
plt.grid(ls="-.",c='k')
plt.show()
```

# **Chapter-10**

```python
import matplotlib.pyplot as plt
import numpy as np 

moneys = np.array([20,10,11,24])
plt.pie(moneys)
plt.grid(ls="-.",c='k')
plt.show()
```

##### *Applying labels and changing angle*

```python
import matplotlib.pyplot as plt
import numpy as np 

moneys = np.array([20,10,11,24])
labelsSS = ["dollars","rupees","euros","notes"]
# Changing angle
plt.pie(moneys,labels=labelsSS,startangle=90)
plt.grid(ls="-.",c='k')
plt.show()
```

#### Exploding chart

```python
import matplotlib.pyplot as plt
import numpy as np 

moneys = np.array([20,10,11,24])
labelsSS = ["dollars","rupees","euros","notes"]
explodes = [0.2,0,0,0.1]
plt.pie(moneys,labels=labelsSS,explode=explodes)
plt.grid(ls="-.",c='k')
plt.show()
```

#### Applying shadow

```python
import matplotlib.pyplot as plt
import numpy as np 

moneys = np.array([20,10,11,24])
labelsSS = ["dollars","rupees","euros","notes"]
explodes = [0.2,0,0,0.1]
plt.pie(moneys,labels=labelsSS,explode=explodes,shadow=True)
plt.grid(ls="-.",c='k')
plt.show()
```

##### Changing colors

```python
import matplotlib.pyplot as plt
import numpy as np 

moneys = np.array([20,10,11,24])
labelsSS = ["dollars","rupees","euros","notes"]
myColors = ["#121212","red","blue","orange"]

plt.pie(moneys,labels=labelsSS,shadow=True,colors=myColors)
plt.grid(ls="-.",c='k')
plt.show()
```

##### Creating legend

```python
import matplotlib.pyplot as plt
import numpy as np 

moneys = np.array([20,10,11,24])
labelsSS = ["dollars","rupees","euros","notes"]
myColors = ["#121212","red","blue","orange"]

plt.pie(moneys,labels=labelsSS,shadow=True,colors=myColors)
plt.legend(title="Currencies")
plt.grid(ls="-.",c='k')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

catogaries = ["cat1","cat2","cat3","cat4"]
ages = [34,21,27,17]
plt.bar(catogaries,ages,color='cyan')
plt.title("Ages Group")
plt.xlabel("catogaries")
plt.ylabel("Ages")

plt.grid(ls="-.",c='k')
plt.show()
```

#### Putting values on bars

```python
import matplotlib.pyplot as plt
import numpy as np 

catogaries = ["cat1","cat2","cat3","cat4"]
ages = [34,21,27,17]
plt.bar(catogaries,ages,color='cyan')
for i,value in enumerate(ages):
    # eachbar , putValue , in string format , horizontal position , vertical position
    plt.text(i,value,str(value),ha='center',va='center')
    
plt.title("Ages Group")
plt.xlabel("catogaries")
plt.ylabel("Ages")

plt.grid(ls="-.",c='k')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np 

catogaries = ["cat1","cat2","cat3","cat4"]
ages = [34,21,27,17]
plt.bar(catogaries,ages,color='cyan')
for i,value in enumerate(ages):
    # eachbar , putValue , in string format , horizontal position , vertical position
    plt.text(i,value-10,str(value),ha='center',va='center')
    
plt.title("Ages Group")
plt.xlabel("catogaries")
plt.ylabel("Ages")

plt.grid(ls="-.",c='k')
plt.show()
```

#### On line chart

```python
import matplotlib.pyplot as plt
import numpy as np 

catogaries = [1,2,3,4]
ages = [34,21,27,17]
plt.plot(catogaries,ages,color='cyan',marker='o')
# Use zip where pair exist
for i,value in zip(catogaries,ages):
    # eachbar , putValue , in string format , horizontal position , vertical position
    plt.text(i,value,str(value),ha='center',va='center')
    
plt.title("Ages Group")
plt.xlabel("catogaries")
plt.ylabel("Ages")

plt.grid(ls="-.",c='k')
plt.show()
```
